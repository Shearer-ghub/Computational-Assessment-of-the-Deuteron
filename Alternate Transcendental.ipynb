```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize as opt
from scipy.optimize import minimize
%matplotlib inline
#x0 is a parameter chosen for known values of a, m, and V0
x0 = 8

x1 = np.linspace(0,4*np.pi,1000)
xp = np.sqrt((x0/x1)**2 - 1)
x = np.linspace(0,20,1000)

plt.figure(figsize=(16,8))
plt.style.use('bmh')

plt.ylim(0,10)
plt.plot(x,np.tan(x),label='$tan(x)$')
plt.plot(x,xp,'r', label='$((x0/x)^2-1)^{1/2}$')
plt.legend(loc=1)

idx = np.argwhere(np.diff(np.sign(np.tan(x) - xp))).flatten()
plt.plot(x[idx], xp[idx], 'ro')
print(xp[idx][0])
print(x[idx][0])
plt.savefig('Trans roots alt.png',bbox_inches = 'tight')
```

```python
m = 1877
hbar = (6.5821e-16)
a = 2.1e-17

xn = x[idx[0]]
E = (((xn*hbar/a)**2)-((8*hbar/a)**2))/2*m
print(E/1e6)
```
